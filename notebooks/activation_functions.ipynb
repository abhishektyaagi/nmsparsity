```python
import torch
import numpy as np

class Net(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1=torch.nn.Linear(in_features=100,out_features=100,bias=True)
        self.fc2=torch.nn.Linear(in_features=3,out_features=1, bias=True)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        return x

net = Net()
fc1_W=net.fc1.weight.detach().numpy()
np.var(fc1_W)*100
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def sigmoid(x):
    return 1/(1+math.e**-x)

def tanh(x):
    return math.tanh(x)

def softsign(x):
    return x/(1+abs(x))


X=np.arange(-4,4,0.1)

act_funcs={"sigmoid": sigmoid, "tanh": tanh, "softsign": softsign}
act_values = {k: None for k in act_funcs.keys()}


for name, activation in act_funcs.items():
    act_values[name] = [activation(x) for x in X]
```

```python
fig,ax=plt.subplots(figsize=(12,8))
for activation_name in act_values:
    ax.plot(X, act_values[activation_name], label=activation_name)
ax.plot([0,0],[-1,1], "--")
plt.legend(loc="upper left")
fig.patch.set_facecolor("white")
fig.savefig("activations.png", bbox_inches="tight")
# fig.bbox_inches("tight")
```
